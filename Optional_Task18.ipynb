```python
print(df.describe())
print("\nAGE")
print("Minimum:",df["age"].min())
print("Maximum:",df["age"].max())
print("Mean:",round(df["age"].mean(),2))
print("Median :",df["age"].median())

print("\nBMI")
print("Minimum:",df["bmi"].min())
print("Maximum:",df["bmi"].max())
print("Mean:",round(df["bmi"].mean(),2))
print("Median:",df["bmi"].median())

print("\nCHILDREN")
print("Minimum:",df["children"].min())
print("Maximum:",df["children"].max())
print("Mea:",round(df["children"].mean(),2))
print("Median:",df["children"].median())

print("\nEXPENSES")
print("Minimum:",df["expenses"].min())
print("Maximum:",df["expenses"].max())
print("Mean:",round(df["expenses"].mean(),2))
print("Median:",df["expenses"].median())
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df["age"],bins=20,kde=True)
plt.title("Age Distribution")
plt.subplot(2,2,2)
sns.histplot(df["bmi"],bins=20,kde=True)
plt.title("BMI Distribution")
plt.subplot(2,2,3)
sns.histplot(df["children"],bins=10,kde=True)
plt.title("Children Distribution")
plt.subplot(2,2,4)
sns.histplot(df["expenses"],bins=20,kde=True)
plt.title("Expenses Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.countplot(x="sex",data=df)
plt.title("Sex")
plt.xticks(rotation=20)
plt.subplot(2,2,2)
sns.countplot(x="smoker",data=df)
plt.title("Smoker")
plt.xticks(rotation=20)
plt.subplot(2,2,3)
sns.countplot(x="region",data=df)
plt.title("Region")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
corr=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
                                                        # Expenses are positively related to age.
                                                        # BMI has a weak positive relationship with expenses.
                                                        # Children have a very weak relationship with expenses.
                                                        # Higher correlation indicates a stronger relationship.
print("Correlation with Expenses:")
print(corr["expenses"].sort_values(ascending=False))
```
