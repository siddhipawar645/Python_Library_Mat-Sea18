```python
df = pd.read_csv("hotel_bookings.csv")
print("Statistical Summary:")
print(df.describe())
key_cols = ["lead_time","stays_in_weekend_nights","stays_in_week_nights","adr"]
print("\nMinimum,Maximum,Mean,and Median Values:")
for col in key_cols:
    print(f"\n{col.upper()}")
    print("Minimum :",df[col].min())
    print("Maximum :",df[col].max())
    print("Mean    :",round(df[col].mean(),2))
    print("Median  :",df[col].median())
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df["lead_time"],bins=20,kde=True)
plt.title("Lead Time Distribution")
plt.subplot(2,2,2)
sns.histplot(df["stays_in_weekend_nights"],bins=20,kde=True)
plt.title("Weekend Nights Distribution")
plt.subplot(2,2,3)
sns.histplot(df["stays_in_week_nights"],bins=20,kde=True)
plt.title("Week Nights Distribution")
plt.subplot(2,2,4)
sns.histplot(df["adr"],bins=20,kde=True)
plt.title("ADR Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.countplot(x="hotel",data=df)
plt.title("Hotel Type")
plt.xticks(rotation=20)
plt.subplot(2,2,2)
sns.countplot(x="meal",data=df)
plt.title("Meal Type")
plt.xticks(rotation=45)
plt.subplot(2,2,3)
sns.countplot(x="market_segment",data=df)
plt.title("Market Segment")
plt.xticks(rotation=45)
plt.subplot(2,2,4)
sns.countplot(x="customer_type",data=df)
plt.title("Customer Type")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

```

```python
corr = df[["lead_time","stays_in_week_nights","stays_in_weekend_nights","adr"]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
                                                            # Lead Time has a weak correlation with ADR.
                                                            # Week Nights and Weekend Nights show a positive correlation.
                                                            # ADR has weak to moderate correlation with the other numeric features.
                                                            # Higher correlation values indicate a stronger relationship.
print("Correlation with ADR:")
print(corr["adr"].sort_values(ascending=False))
```
