```python
print("1.Price Statistics")
print("Minimum:",df["price"].min())
print("Maximum:",df["price"].max())
print("Mean:",df["price"].mean())
print("Median:",df["price"].median())

print("\n2.Mileage Statistics")
print("Minimum:",df["mileage"].min())
print("Maximum:",df["mileage"].max())
print("Mean:",df["mileage"].mean())
print("Median:",df["mileage"].median())

print("\n3.Year Statistics")
print("Minimum:",df["year"].min())
print("Maximum:",df["year"].max())
print("Mean:",df["year"].mean())
print("Median:",df["year"].median())


```

```python
plt.subplot(2,3,1)
sns.histplot(df["price"],bins=20,kde=True)
plt.title("Price Distribution")
plt.subplot(2,3,2)
sns.histplot(df["mileage"], bins=20,kde=True)
plt.title("Mileage Distribution")
plt.subplot(2,3,3)
sns.histplot(df["year"], bins=20,kde=True)
plt.title("Year Distribution")
plt.subplot(2,3,4)
sns.histplot(df["engineSize"], bins=20,kde=True)
plt.title("Engine Size Distribution")
plt.subplot(2,3,5)
sns.histplot(df["mpg"], bins=20,kde=True)
plt.title("MPG Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,1)
sns.countplot(x="fuelType",data=df)
plt.title("Fuel Type")
plt.xticks(rotation=20)
plt.subplot(2,2,2)
sns.countplot(x="transmission",data=df)
plt.title("Transmission")
top10=df["model"].value_counts().head(10).index
plt.figure(figsize=(4,6))
sns.countplot(x="model", data=df,order=top10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
corr=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()                                           # Price is positively related to year and engine size.
                                                     # Price is negatively related to mileage.
                                                    # Higher correlation indicates a stronger relationship.
print("Correlation with Price:")
print(corr["price"].sort_values(ascending=False))
```

```python
df = pd.read_csv("ford_car_dataset.csv")
print("Missing Values:")
print(df.isnull().sum())
df = df.drop_duplicates()
num_cols = ["price", "mileage", "year", "engineSize", "mpg"]
plt.figure(figsize=(15,10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2,3,i)
    sns.histplot(df[col], bins=20)
    plt.title(col)
plt.tight_layout()
plt.show()
plt.figure(figsize=(18,10))
plt.subplot(2,2,1)
sns.countplot(x="fuelType", data=df)
plt.xticks(rotation=45)
plt.subplot(2,2,2)
sns.countplot(x="transmission", data=df)
plt.xticks(rotation=45)
plt.subplot(2,2,3)
sns.countplot(y="model", data=df,
              order=df["model"].value_counts().index)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
X = df.drop("price", axis=1)
y = df["price"]
print("\nIndependent Features:")
print(X.columns)
print("\nDependent Feature:")
print(y.name)
df_encoded = pd.get_dummies(df, columns=df.select_dtypes(include="object").columns, dtype=int)
print("\nEncoded Data:")
print(df_encoded.head())
X = df_encoded.drop("price", axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print("\nFirst 5 Rows of Scaled Data:")
print(X_scaled.head())
```
