```python
print(df.describe())
print("\nAGE")
print("Minimum:",df["age"].min())
print("Maximum:",df["age"].max())
print("Mean:",round(df["age"].mean(),2))
print("Median:",df["age"].median())

print("\nHEIGHT_CM")
print("Minimum:",df["height_cm"].min())
print("Maximum:",df["height_cm"].max())
print("Mean:",round(df["height_cm"].mean(),2))
print("Median:",df["height_cm"].median())

print("\nWEIGHT_KG")
print("Minimum:",df["weight_kg"].min())
print("Maximum:",df["weight_kg"].max())
print("Mean:",round(df["weight_kg"].mean(),2))
print("Median:",df["weight_kg"].median())

print("\nMARKET_VALUE_EUR")
print("Minimum:",df["market_value_eur"].min())
print("Maximum:",df["market_value_eur"].max())
print("Mean:",round(df["market_value_eur"].mean(),2))
print("Median:",df["market_value_eur"].median())
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df["age"],bins=20,kde=True)
plt.title("Age Distribution")
plt.subplot(2,2,2)
sns.histplot(df["height_cm"],bins=20,kde=True)
plt.title("Height Distribution")
plt.subplot(2,2,3)
sns.histplot(df["weight_kg"],bins=20,kde=True)
plt.title("Weight Distribution")
plt.subplot(2,2,4)
sns.histplot(df["market_value_eur"],bins=20,kde=True)
plt.title("Market Value Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.countplot(x="nationality",data=df)
plt.title("Nationality")
plt.xticks(rotation=90)
plt.subplot(2,2,2)
sns.countplot(x="preferred_foot",data=df)
plt.title("Preferred Foot")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
corr = df[["age","height_cm","weight_kg"]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
                                    # Age has a weak correlation with Height and Weight.
                                    # Height and Weight show a positive correlation.
                                    # Higher correlation indicates a stronger relationship.
print("Correlation with Age:")
print(corr["age"].sort_values(ascending=False))
```

```python
X=df.drop("age",axis=1)
print("Independent Features:")
print(X.columns)
y=df["age"]
print("\nDependent Feature:")
print(y.name)
```

```python
cat_cols=df.select_dtypes(include="object").columns

print("Categorical Columns:")
print(cat_cols)

print("\nBefore Encoding:")
print(df[["preferred_foot", "position"]].head())

df_encoded=pd.get_dummies(df, columns=cat_cols,dtype=int)

print("\nAfter Encoding:")
print(df_encoded.head())

print("\nShape after Encoding:",df_encoded.shape)
```

```python
from sklearn.preprocessing import StandardScaler
X=df_encoded.drop("age",axis=1)
X_numeric=X.select_dtypes(include=["int64", "float64"])
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_numeric)
X_scaled=pd.DataFrame(X_scaled,columns=X_numeric.columns)
# Display first 5 rows
print("First 5 Rows of Scaled Data:")
print(X_scaled.head())
```
